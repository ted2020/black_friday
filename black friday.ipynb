```r
library(dplyr)
library(tidyverse)
library(tidytext)
library(ggplot2)
```

```r
# https://www.finder.com/black-friday-statistics
```

```r
blackfriday <- read.csv("BlackFriday.csv")
```

```r
head(blackfriday)
```

```r
head(blackfriday %>% group_by(City_Category) %>% filter(Gender=="F") %>% arrange(desc(Purchase)))
```

```r
blackfriday <- blackfriday %>% mutate(gender_numeric = as.numeric(as.factor(blackfriday$Gender)))
# 1 for F, 2 for M
```

```r
cor(blackfriday$gender_numeric,blackfriday$Purchase)
```

```r
levels(blackfriday$Stay_In_Current_City_Years)
```

```r
blackfriday <- blackfriday %>% mutate(stayincurrentcity_numeric = as.numeric(as.factor(blackfriday$Stay_In_Current_City_Years)))
```

```r
head(blackfriday)
```

```r
sum(blackfriday$Marital_Status)/length(blackfriday$Marital_Status)
```

```r
blackfriday %>% group_by(gender_numeric) %>% summarize(sum(Purchase/n()))
```

```r
blackfriday %>% group_by(Marital_Status) %>% summarize(sum(Purchase/n()))
```

```r
blackfriday %>% group_by(stayincurrentcity_numeric) %>% summarize(sum(Purchase/n()))
```

```r
blackfriday <- blackfriday %>% mutate_if(is.integer, ~replace(., is.na(.), 0))
```

```r
head(blackfriday)
# product category 0 means NA
```

```r
unique(blackfriday$Product_Category_1)
unique(blackfriday$Product_Category_2)
unique(blackfriday$Product_Category_3)

```

```r
blackfriday %>% group_by(Product_Category_1) %>% filter(Gender=="F") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
```

```r
blackfriday %>% group_by(Product_Category_1) %>% filter(Gender=="M") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
```

```r
blackfriday %>% group_by(Product_Category_1) %>% filter(Marital_Status==0) %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
```

```r
blackfriday %>% group_by(Product_Category_1) %>% filter(Marital_Status==1) %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
```

```r
# looks like certain product categories are more appealing to everyone, whether male or female or single or married.
```

```r
blackfriday %>% group_by(Occupation) %>% filter(Gender=="F") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
```

```r
blackfriday %>% group_by(Occupation) %>% filter(Gender=="M") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
```

```r
# though male shoppers tend to average out roughly the same across shopping categories, 
# female shoppers tend to spend a bit more on certain categories
```

```r
levels(blackfriday$Age)
class(blackfriday$Age)
```

```r
blackfriday %>% group_by(Age,Occupation) %>% filter(Gender=="M") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))

# different age groups focus on different shopping categories
# shopping preferences change with the gender. female and male shopping behavior change with age.
```

```r
blackfriday %>% group_by(Age,Occupation) %>% filter(Gender=="F") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
```

```r
blackfriday %>% group_by(City_Category) %>% filter(Gender=="M") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
blackfriday %>% group_by(City_Category) %>% filter(Gender=="F") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))

# both male and female from city C spend more
```

```r
blackfriday %>% group_by(City_Category,Product_Category_1) %>% filter(Gender=="M") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))

# products 10, 6, and 7 from category 1 are more preferred by male
```

```r
blackfriday %>% group_by(City_Category,Product_Category_1) %>% filter(Gender=="F") %>%
     summarize(avg=sum(Purchase/n())) %>% arrange(desc(avg))
```
